```python
import sys
sys.path.append('../code/')

from facenet_keras_v2 import *
model = InceptionResNetV1(weights_path='../model/facenet-keras/weights/facenet_keras_weights.h5')
```

```python
from coremltools.proto import NeuralNetwork_pb2

# The conversion function for Lambda layers.
def convert_lambda(layer):
    # Only convert this Lambda layer if it is for our swish function.
    if layer.function == scaling:
        print("this is scaling!!!!!!!!")
        params = NeuralNetwork_pb2.CustomLayerParams()

        # The name of the Swift or Obj-C class that implements this layer.
        params.className = "scaling"

        # The desciption is shown in Xcode's mlmodel viewer.
        params.description = "scaling input"
        print("arguments:",layer.arguments)
        
        params.parameters["scale"].doubleValue = layer.arguments['scale']
        # Set configuration parameters
        #params.parameters["someNumber"].intValue = 100
        # params.parameters["someString"].stringValue = "Hello, world!"
        
        # Add some random weights
        # my_weights = params.weights.add()
        # my_weights.floatValue.extend(np.random.randn(10).astype(float))
        
        

        return params
    else:
        return None
```

```python
def convert_lambda_new(layer):
    if layer.function == scaling:
        print("this is scaling!!!!!!!!")
        params = NeuralNetwork_pb2.CustomLayerParams()
        params.className = "Scaling"
        params.description = "scaling input"
        params.parameters["scale"].doubleValue = layer.arguments['scale']
        return params
    else:
        return None
```

```python
#(6) convert_lambda_new
coreml_model = coremltools.converters.keras.convert(
    model,
    input_names="image",
    image_input_names="image",
    output_names="output",
    add_custom_layers=True,
    custom_conversion_functions={ "Lambda": convert_lambda_new })
```

```python
# convert_lambda_new
print(coreml_model)
coreml_model.save('FaceNet6.mlmodel')
```

```python
#(7) convert_lambda_new , image_scale 1/255
coreml_model = coremltools.converters.keras.convert(
    model,
    input_names="image",
    image_input_names="image",
    output_names="output",
#     red_bias=-1,
#     green_bias=-1,
#     blue_bias=-1,
    image_scale=1.0/255.0,
    add_custom_layers=True,
    custom_conversion_functions={ "Lambda": convert_lambda_new })
```

```python
print(coreml_model)
coreml_model.save('FaceNet7.mlmodel')
```

```python
#(8) convert_lambda_new , image_scale 2/255, red_bias
coreml_model = coremltools.converters.keras.convert(
    model,
    input_names="image",
    image_input_names="image",
    output_names="output",
    red_bias=-1,
    green_bias=-1,
    blue_bias=-1,
    image_scale=2.0/255.0,
    add_custom_layers=True,
    custom_conversion_functions={ "Lambda": convert_lambda_new })
```
