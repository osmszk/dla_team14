````python
from PIL import Image  
from matplotlib.pyplot import imshow

img_clapton1 = Image.open(img_path+file1) 

%matplotlib inline
imshow(np.asarray(img_clapton1))
````

````python
predictions_clapton1 = model.predict({"data":img_clapton1})
rep_clapton1 = predictions_clapton1['output']
print rep_clapton1
````

````python
img2 = Image.open(img_path+file2) 
imshow(np.asarray(img2))
predictions_clapton2 = model.predict({"data":img2})
rep_clapton2 = predictions_clapton2['output']
print rep_clapton2
````

````python
img3 = Image.open(img_path+file3) 
imshow(np.asarray(img3))
predictions_lennon2 = model.predict({"data":img3})
rep_lennon2 = predictions_lennon2['output']
print rep_lennon2
````

````python
#(1)clapton-1とclapton-2
d1 = rep_clapton1 - rep_clapton2
dot1 = np.dot(d1, d1)
print dot1
print np.sqrt(dot1)
````

````python
#(2)clapton-1とlennon-2
d2 = rep_clapton1 - rep_lennon2
dot2 = np.dot(d2, d2)
print dot2
print np.sqrt(dot2)
````

````python
#(3)clapton-2とlennon-2
d3 = rep_clapton2 - rep_lennon2
dot3 = np.dot(d3, d3)
print dot3
print np.sqrt(dot3)
````
